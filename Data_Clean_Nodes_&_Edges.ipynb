```python
# Putting all zipped edges columns into new separate dataframe

Zipped = final[['Zip 1','Zip 2','Zip 3','Zip 4','Zip 5','Zip 6','Zip 7','Zip 8','Zip 9','Zip 10','Zip 11','Zip 12','Zip 13','Zip 14','Zip 15','Zip 16','Zip 17','Zip 18','Zip 19','Zip 20','Zip 21','Zip 22','Zip 23']].copy()
Zipped
```

```python
# Stacking all edges columns into one big long column in a new dataframe
Node_Edge = Zipped.stack().reset_index()

# New dataframe where the two user id values are now split into two separate columns
Nodes_Edges = pd.DataFrame(Node_Edge[0].tolist(), index=Node_Edge.index)
Nodes_Edges
```

## 3.) Appending engagement data to the nodes list 
#### Necessary to create a color map for network visualizations later on related to engagement levels

```python
# Realized later on we needed to link extra information about each user to the node User_ID where possible
# Some users will not have all information because they only appear as mentioned user, not as a primary user

# Pulled identifying information from before and replies_count, retweets_count, and likes_count from original data
# Want to sum the total engagement a user has received in the dataset to get a sense of which nodes have more influence and reach on Twitter
# Can then use color to classify nodes in visualizations later on

Engagement_Data = pd.DataFrame().assign(id=df['id'], conversation_id=df['conversation_id'], user_id=df['user_id'], username=df['username'], replies_count=df['replies_count'], retweets_count=df['retweets_count'], likes_count=df['likes_count'])
print(Engagement_Data)
```

```python
# Pulling just user_id and replies_count from engagement-related dataframe

Replies_Data = pd.DataFrame().assign(user_id=Engagement_Data['user_id'], replies_count=Engagement_Data['replies_count'])

# Obtaining unique list of users with the total replies they received across the dataset summed
# Goal to extract a measure of engagement with their account from the day this data was captured

Replies_Summed = Replies_Data.groupby('user_id')['replies_count'].sum().reset_index()
Replies_Summed
```

```python
# Double checking that the groupby method worked and that all replies to that account in the data set are correctly summed
# Did work, checked CNN's as it had multiple tweets in the dataset

Replies_Summed.loc[Replies_Summed['user_id'] == 759251]
```

```python
# Pulled just user_id and retweets_count from engagement-related dataframe

Retweets_Data = pd.DataFrame().assign(user_id=Engagement_Data['user_id'], retweets_count=Engagement_Data['retweets_count'])

# Grouping by sum of retweets user received
Retweets_Summed = Retweets_Data.groupby('user_id')['retweets_count'].sum().reset_index()
Retweets_Summed
```

```python
# Double checking that the groupby method worked and that all retweets to that account in the data set are correctly summed
# Did work, checked BBC's as it had multiple tweets in the dataset

Retweets_Summed.loc[Retweets_Summed['user_id'] == 612473]
```

```python
# Pulled just user_id and likes_count from engagement-related dataframe

Likes_Data = pd.DataFrame().assign(user_id=Engagement_Data['user_id'], likes_count=Engagement_Data['likes_count'])

# Grouping by sum of likes user received
Likes_Summed = Likes_Data.groupby('user_id')['likes_count'].sum().reset_index()
Likes_Summed
```

```python
# Double checking that the groupby method worked and that all likes to that account in the data set are correctly summed
# Did work, checked BBC's as it had multiple tweets in the dataset

Likes_Summed.loc[Likes_Summed['user_id'] == 612473]
```

```python
# Need to connect the summed replies, retweets, likes to the corresponding user ids for future network graphs

# Loaded in the cleaned list of unique nodes with Screen name attached
Nodes = pd.read_csv ('Mentions_Nodes.csv')
Nodes

# Renamed columns to make them easier to call later on
column_name_update_map = {'User ID':'user_id'}
Nodes = Nodes.rename(columns=column_name_update_map) 
Nodes
```

```python
# Merging the grouped by columns with counts of replies, retweets, and likes to dataframe with id and screen name
# Merged on user id to correctly connect the data

Nodes_Merge_1 = pd.merge(Nodes, Replies_Summed, on='user_id', how='outer')
Nodes_Merge_2 = pd.merge(Nodes_Merge_1, Retweets_Summed, on='user_id', how='outer')
Nodes_Merge_3 = pd.merge(Nodes_Merge_2, Likes_Summed, on='user_id', how='outer')
Nodes_Merge_3
```

```python
# Filling the null cells in the counted columns with zeros
# Nulls came from users that were only mentioned in the data set so no engagement information was available

Nodes_Merge_3['replies_count'] = Nodes_Merge_3['replies_count'].fillna(0)
Nodes_Merge_3['retweets_count'] = Nodes_Merge_3['retweets_count'].fillna(0)
Nodes_Merge_3['likes_count'] = Nodes_Merge_3['likes_count'].fillna(0)
Nodes_Merge_3
```

```python
# Making the count columns into the integer data type 

Nodes_Merge_3 = Nodes_Merge_3.astype({"replies_count":"int","retweets_count":"int","likes_count":"int"})
Nodes_Merge_3
```
